```python
# Pasar el archivo a un archivo de texto para luego procesarlo, mismo proceso que con el df reviews

with gzip.open(r'C:\Users\user\OneDrive\Escritorio\SOYHENRY\Curso Data Science\PI1\steam_games.json.gz', 'rb') as f_in:
    with open('C:\\Users\\user\\OneDrive\\Escritorio\\SOYHENRY\\Curso Data Science\\PI1\\steam_games.txt', 'wb') as f_out:
        f_out.write(f_in.read())
```

```python
#Cambiar la ruta del archivo
file_path = r'C:\Users\user\OneDrive\Escritorio\SOYHENRY\Curso Data Science\PI1\steam_games.txt'
# Función para cargar el archivo de texto y convertirlo en una lista de diccionarios
def load_data_to_dict(file_path):
    json_data = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line_number, line in enumerate(file, start=1):
            line = line.strip()

            # Reemplazar "NaN" con None (nulo en Python)
            line = re.sub(r'\bNaN\b', 'None', line)

            # Corregir dobles comillas incorrectas en URLs
            line = re.sub(r'""http', '"http', line)

            # Reemplazar "true" por "True" y "false" por "False"
            line = re.sub(r'\btrue\b', 'True', line)
            line = re.sub(r'\bfalse\b', 'False', line)

            try:
                # Evaluar la línea como un diccionario de Python usando eval
                parsed_line = eval(line)
                json_data.append(parsed_line)
            except Exception as e:
                print(f"Error parsing line {line_number}: {line}\nError: {e}")

    return json_data

# Ruta del archivo de texto
output_txt_path = 'C:/Users/user/OneDrive/Escritorio/SOYHENRY/Curso Data Science/PI1/steam_games.txt'

# Cargar los datos y convertirlos en un DataFrame
json_data = load_data_to_dict(output_txt_path)

# Convertir la lista de diccionarios en un DataFrame de pandas
dfgames = pd.DataFrame(json_data)

# Imprimir el DataFrame
print(dfgames)
```

```python
#Crear un analizador de sentimientos
analyzer = SentimentIntensityAnalyzer()

# Función para el análisis de sentimiento usando VADER
def analyze_sentiment_vader(row):
    review = row['review']
    recommend = row['recommend']
    
    if pd.isna(review):
        return 1  # Neutral para valores nulos
    
    # Si recommend es False, asignamos automáticamente un sentimiento negativo (0)
    if recommend == False:
        return 0  # Negativo si no recomienda el producto
    
    # Caso contrario, realizamos el análisis de sentimiento
    score = analyzer.polarity_scores(review)
    
    # Calcular el sentimiento basado en el análisis
    if score['compound'] > 0:
        return 2  # Positivo
    elif score['compound'] == 0:
        return 1  # Neutral
    else:
        return 0  # Negativo

dfreviews['sentiment_analysis'] = dfreviews.apply(analyze_sentiment_vader, axis=1)
```
